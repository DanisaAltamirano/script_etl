```python
import os
import pandas as pd
```

```python
# Ruta de la carpeta que contiene los archivos CSV
directorio_csv = "/content/sample_data"

# Lista para almacenar los datos de los archivos CSV
datos_csv = []

# Itera a través de los archivos en la carpeta
for archivo in os.listdir(directorio_csv):
    if archivo.endswith(".csv"):
        ruta_archivo = os.path.join(directorio_csv, archivo)
        # Carga el archivo CSV en un DataFrame de pandas
        df = pd.read_csv(ruta_archivo)
        datos_csv.append((archivo, df))

# Ahora, la lista 'datos_csv' contiene tuplas donde el primer elemento es el nombre del archivo y el segundo es un DataFrame de pandas con los datos del archivo CSV.


#Imprimir la información de cada archivo CSV
for nombre_archivo, dataframe in datos_csv:
    print(f"Nombre del archivo: {'california_housing_test'}")
    print("Contenido del archivo CSV:")
    print(dataframe)
    print("\n")
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

# Agrupa los datos y cuenta la cantidad de elementos en cada grupo
conteo_por_severidad = columnas_filtradas.groupby('Severity Level').size()

# Ordena el DataFrame en orden descendente según la cantidad
conteo_por_severidad = conteo_por_severidad.sort_values(ascending=True)

# Crea un gráfico de barras horizontales
ax = conteo_por_severidad.plot(kind='barh', color=sns.color_palette('Dark2'))

# Oculta los bordes superior y derecho del gráfico
plt.gca().spines[['top', 'right']].set_visible(False)

# Agrega las cantidades (números) a las barras
for i, v in enumerate(conteo_por_severidad):
    ax.text(v + 1, i, str(v), color='black', va='center')

# Establece una etiqueta para el eje y
plt.xlabel('Cantidad')

# Muestra el gráfico
plt.show()

```
